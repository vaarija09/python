<b> Class <b>

```python
class car:

    color = ""
    enginetype =""

    def __init__(self, color,enginetype):
        self.color = color
        self.enginetype = enginetype

    def get_color(self):
        print("color--->", self.color)

    def get_enginetype(self):
        print("engine---->", self.enginetype)
```

```python
car1 = car("red","petrol")  #object created
car1.get_color()
car1.get_enginetype()
```

```python
car2 = car("green","disel")  #object created
car2.get_color()
car2.get_enginetype()
```
