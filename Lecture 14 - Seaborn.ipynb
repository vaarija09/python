<b> Viewing statistical releations<b>

```python
#scatter plots
#line plots
```

```python
sns.scatterplot(data=tips, x = tips["sex"], y = tips["tip"])
```

```python
sns.scatterplot(data=tips, x = "sex", y = "tip")
```

```python
sns.scatterplot(data=tips, x = "total_bill", y = "tip")
```

```python
sns.lineplot(data = tips, x = "total_bill", y = "tip")
```

```python
sns.relplot(data = tips, x = "total_bill", y = "tip")
```

```python
sns.relplot(data = tips, x = "total_bill", y = "tip", kind="line")
```

```python
sns.relplot(data = tips, x = "total_bill", y = 'tip', hue="time")
```

```python
sns.relplot(data = tips, x = "total_bill", y = "tip", hue="time", style="smoker")
```

```python
sns.relplot(data = tips, x = "total_bill", y = "tip", hue="time", col="day")
```

```python
sns.relplot(data = tips[tips.sex == "Male" ], x = "total_bill", y = "tip", hue="time")
```

```python
sns.relplot(data = tips[tips.sex == "Female" ], x = "total_bill", y = "tip", hue="time")
```

<b> Univariate plots<b>

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.info()
```

```python
titanic.head(2)
```

```python
sns.distplot(titanic["survived"])
```

```python
sns.distplot(titanic["survived"], kde=False)
```

```python
sns.distplot(titanic["survived"], hist=False)
```

```python
import matplotlib.pyplot as plt
```

```python
sns.distplot(titanic["survived"], label="survived")
sns.distplot(titanic["pclass"], label="pclass")
plt.legend()
```

<b>Bivariate analysis <b>

```python
sns.jointplot(x = titanic.pclass, y = titanic.age, data = titanic)
```
