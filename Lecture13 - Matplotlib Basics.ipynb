```python
fig, ax = plt.subplots()

index = bikes.season.unique()
bar_width = 0.5 
values = df_by_seasons.cnt.sum()

ax.bar(index, values, bar_width)

```

```python
fig, ax = plt.subplots()
ax.scatter(bikes.casual, bikes.registered)
plt.show()
```

```python
fig, ax = plt.subplots()

values = df_by_seasons.cnt.sum()
labels = [ "s1", "s2", "s3", "s4"]
#labels = [ 1, 2, 3, 4]
ax.pie(values, labels=labels)

plt.show()
```

```python
list1 = ["R1", "R2", "R3", "R4"]
list2 = [100, 200, 150, 175]


plt.plot(list1, list2, label="Sales")

plt.title("Sales Over Regions")
plt.xlabel("Regions")
plt.ylabel("Sales")
plt.legend()
```

```python
list1 = ["R1", "R2", "R3", "R4"]
list2 = [100, 200, 150, 175]
list3 = [10, 20, 15, 17.5]

plt.plot(list1, list2, label="Sales")
plt.plot(list1, list3, label="Cost")

plt.title("Sales Over Regions")
plt.xlabel("Regions")
plt.ylabel("Sales")
plt.legend()
```
